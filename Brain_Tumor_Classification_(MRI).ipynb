# **Pengenalan Tentang Tumor Otak**
---
**Tumor otak** adalah pertumbuhan abnormal sel di dalam otak yang bisa bersifat jinak (non-kanker) atau ganas (kanker). Tumor ini dapat berasal dari jaringan otak itu sendiri (disebut sebagai tumor otak primer) atau menyebar dari bagian tubuh lain (disebut tumor otak sekunder atau metastasis). Tumor otak primer yang paling umum termasuk glioma, meningioma, dan glioblastoma, dengan glioblastoma menjadi salah satu jenis yang paling agresif.

#### **Jenis-jenis Tumor Otak:**
1. **Glioma:** Ini adalah kelompok tumor yang berasal dari sel glial yang menyokong sistem saraf. Jenis glioma termasuk astrositoma, oligodendroglioma, dan ependimoma. Glioblastoma multiforme (GBM) adalah bentuk paling agresif dari glioma dengan prognosis yang buruk dan harapan hidup rata-rata 12-16 bulan setelah diagnosis, meskipun mendapat terapi intensif.

2. **Meningioma:** Biasanya tumor jinak yang berkembang dari meninges (lapisan pelindung otak dan sumsum tulang belakang). Ini adalah jenis tumor otak primer yang paling umum pada orang dewasa.

3. **Tumor metastasis:** Ini berasal dari kanker di bagian tubuh lain (seperti paru-paru atau payudara) yang menyebar ke otak. Tumor jenis ini lebih umum daripada tumor otak primer.

#### **Gejala Tumor Otak:**
* Sakit kepala yang semakin parah atau terjadi lebih sering
* Mual dan muntah tanpa sebab yang jelas
* Gangguan penglihatan seperti penglihatan ganda
* Kelemahan atau mati rasa pada salah satu sisi tubuh
* Kesulitan berbicara atau memahami pembicaraan
* Kejang (epilepsi)
* Perubahan kepribadian atau perilaku

#### **Metode Diagnostik:**
Diagnosis tumor otak umumnya melibatkan pencitraan seperti MRI, CT scan, dan PET scan. Teknologi canggih seperti MR spectroscopy dan fMRI dapat membantu dalam membedakan antara tumor jinak dan ganas serta menentukan karakteristik molekuler tumor. Metode terbaru, seperti penggunaan radiomics dan kecerdasan buatan (AI), meningkatkan kemampuan diagnostik dengan menganalisis fitur gambar yang lebih rinci dan memprediksi mutasi genetik tertentu.

Terdapat beberapa kategori atau jenis pemindaian MRI yang sering digunakan untuk mendiagnosis, mengevaluasi, dan mengklasifikasikan tumor otak. Berikut penjelasan tentang kategori **FLAIR, T1, T1CE, T2, dan SEG**.

1. FLAIR (Fluid Attenuated Inversion Recovery)
  * **Apa itu:** Teknik MRI yang memodifikasi sinyal dari cairan agar tidak muncul terlalu terang seperti pada pemindaian T2 biasa. Hal ini dilakukan dengan menekan sinyal dari cairan serebrospinal (CSF).
  * **Fungsi:** Sangat baik dalam menampilkan lesi di area otak yang biasanya tertutup oleh sinyal cairan di T2, seperti gliosis, edema, atau tumor. Membantu mendeteksi lesi di dekat ventrikel otak atau di dalam materi putih.
  * **Aplikasi:** Sering digunakan untuk mendeteksi penyakit demielinisasi (misalnya, multiple sclerosis), tumor otak, atau infark.

2. T1-weighted (T1)
  * **Apa itu:** Jenis MRI yang menghasilkan gambar berdasarkan waktu relaksasi T1. Jaringan dengan waktu relaksasi T1 yang pendek (seperti lemak) akan tampak terang, sementara jaringan dengan waktu relaksasi lebih lama (seperti cairan) akan tampak gelap.
  * **Fungsi:** Sering digunakan untuk anatomi otak karena memberikan kontras yang baik antara materi abu-abu dan materi putih.
  * **Aplikasi:** Digunakan untuk evaluasi anatomi umum otak, deteksi perdarahan akut, dan melihat batas-batas tumor.

3. T1CE (T1-weighted with Contrast Enhancement)
  * **Apa itu:** Gambar T1 yang diambil setelah pemberian agen kontras (seperti gadolinium). Agen kontras ini membantu meningkatkan visibilitas jaringan yang memiliki banyak pembuluh darah atau peningkatan permeabilitas vaskular.
  * **Fungsi:** Sangat penting untuk menilai tumor karena kontras akan menyoroti area dengan peningkatan pembuluh darah abnormal, yang sering kali menunjukkan adanya massa tumor atau area gangguan sawar darah otak (blood-brain barrier).
  * **Aplikasi:** Sering digunakan untuk mendeteksi tumor ganas, abses, infeksi, dan untuk membedakan antara tumor yang aktif dan yang sudah tidak aktif.

4. T2-weighted (T2)
  * **Apa itu:** Jenis MRI yang menonjolkan sinyal berdasarkan waktu relaksasi T2. Cairan (seperti CSF) tampak terang, sedangkan lemak tampak lebih gelap dibandingkan dengan gambar T1.
  * **Fungsi:** Baik untuk mendeteksi edema, cairan, atau lesi yang berisi cairan seperti kista atau edema tumor.
  * **Aplikasi:** Sering digunakan untuk identifikasi edema otak, perubahan patologis di otak, dan deteksi tumor yang memiliki komponen cair.

5. SEG (Segmentation)
  * **Apa itu:** Proses segmentasi pada gambar MRI yang digunakan untuk memisahkan berbagai jenis jaringan atau struktur dalam otak (misalnya, tumor, edema, atau jaringan sehat).
  * **Fungsi:** Memungkinkan analisis kuantitatif volume tumor, edema, atau perubahan struktural lainnya. Biasanya dilakukan dengan menggunakan algoritma machine learning atau deep learning.
  * **Aplikasi:** Digunakan dalam penelitian klinis dan praktik klinis untuk pemetaan tumor yang lebih akurat dan untuk membantu perencanaan bedah atau radioterapi.

#### **Pilihan Pengobatan:**
Pengobatan tumor otak tergantung pada jenis, ukuran, dan lokasi tumor, serta kesehatan umum pasien. Pendekatan pengobatan meliputi:

* **Operasi:** Untuk mengangkat sebanyak mungkin jaringan tumor.
Radioterapi: Menggunakan sinar radiasi untuk membunuh sel kanker yang tersisa.
* **Kemoterapi:** Penggunaan obat untuk membunuh sel kanker, misalnya, temozolomide yang sering digunakan untuk glioblastoma.
* **Terapi Targeted dan Imunoterapi:** Pendekatan baru yang lebih spesifik dengan target molekuler tertentu pada sel kanker.

Penelitian terbaru menunjukkan potensi terapi seperti **CAR T-cell therapy** dan penggunaan **neuroimaging** yang lebih canggih untuk meningkatkan hasil pengobatan pasien dengan tumor otak ganas seperti glioblastoma.

# **Create Modelling**

```python
# Import package yang diperlukan

import tensorflow as tf
import cv2
import numpy as np
from tensorflow import keras
from sklearn import preprocessing
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Conv2D, MaxPooling2D, Dropout, Flatten, Dense, Activation, BatchNormalization
print(tf.__version__)
```

```python
# Create model

model = tf.keras.models.Sequential([
    tf.keras.layers.Conv2D(32, (3, 3), activation='relu', input_shape=(150, 150, 3)),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(64, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Conv2D(128, (3, 3), activation='relu'),
    tf.keras.layers.MaxPooling2D(2, 2),
    tf.keras.layers.Flatten(),
    tf.keras.layers.Dense(512, activation='relu'),
    tf.keras.layers.Dense(4, activation='softmax') # Menggunakan 4 output neurons dan 'softmax' activation
])

# Tampilkan summary
model.summary()
```

```python
# Compile model dengan menggunakan Adam

from tensorflow.keras.optimizers import Adam

model.compile(
    loss='categorical_crossentropy', # Gunakan categorical crossentropy untuk multi-kelas
    optimizer='adam',
    metrics=['accuracy']
)
```

# **Data Processing**

```python
# Set up data generator

from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rescale=1./255)
test_datagen = ImageDataGenerator(rescale=1./255)
```

```python
# Preprocess Data

# Training images
train_generator = train_datagen.flow_from_directory(
    '/content/Training',  # direktori data training
    target_size=(150, 150),  # mengubah resolusi seluruh gambar menjadi 150x150 piksel
    batch_size=20,
    class_mode='categorical'
)

# Validation images
validation_generator = test_datagen.flow_from_directory(
    '/content/Testing',  # direktori data validasi
    target_size=(150, 150),
    batch_size=20,
    class_mode='categorical'
)
```

```python
# Fit model
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // train_generator.batch_size,
    epochs=10,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // validation_generator.batch_size
)
```

```python
from google.colab import files
from keras.preprocessing import image
import numpy as np

# Upload file dari lokal
uploaded = files.upload()

# Loop melalui semua file yang diupload
for fn in uploaded.keys():

    # Memuat dan memproses gambar
    path = '/content/' + fn
    img = image.load_img(path, target_size=(150, 150))
    x = image.img_to_array(img)
    x = x / 255.0  # Normalisasi gambar
    x = np.expand_dims(x, axis=0)

    # Membuat prediksi
    images = np.vstack([x])
    classes = model.predict(images, batch_size=10)
    print(f"Prediksi mentah: {classes[0]}")

    # Menentukan kelas berdasarkan prediksi
    class_names = ['glioma tumor', 'meningioma tumor', 'no tumor', 'pituitary tumor']
    predicted_class = class_names[np.argmax(classes[0])]
    print(f"{fn} is a {predicted_class}")
```
